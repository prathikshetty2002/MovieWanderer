##Step 3: Data Pre processing

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
def helper(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
import ast
```

```python
movies['genres'] = movies['genres'].apply(helper)
```

```python
movies
```

```python
movies['keywords'] = movies['keywords'].apply(helper)
```

```python
def cast(obj):
    L=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=3:
            L.append(i['name'])
            count+=1
        else:
            break
    return L
```

```python
movies['cast']= movies['cast'].apply(cast)
```

```python
movies.head()
```

```python
movies.iloc[0].crew
```

```python
def crew(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
movies['crew']=movies['crew'].apply(crew)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python

movies['cast'] = movies['cast'].apply(lambda x:[ i.replace(' ','') for i in x])
movies['genres'] = movies['genres'].apply(lambda x:[ i.replace(' ','') for i in x])
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x: [ i.replace(' ','') for i in x])
```

```python
movies['crew'] = movies['crew'].apply(lambda x: [ i.replace(' ','') for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['genres'] + movies['overview'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
movies_df = movies[['id','title','tags']]
movies_df.head()
```

```python
movies_df['tags'] = movies_df['tags'].apply(lambda x:' '.join(x))
```

```python
movies_df
```

```python
movies_df['tags'][0]
```

```python
movies_df['tags'] = movies_df['tags'].apply(lambda x:x.lower())
```

```python
movies_df
```

##Step 4: Vectorization - to convert text to vector

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(movies_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

## Step 5: NLP(Natural language processing)
#stemming - to convert words to their root word 

Eg : Dance, Dancing, Danced - Dance 

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
 
```

```python
movies_df['tags'].apply(stem)
```

```python
movies_df['tags'].iloc[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index=movies_df[movies_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list= sorted(list(enumerate(distances)),reverse = True, key = lambda x:x[1])[1:6]
    for i in movies_list:
        print(movies_df.iloc[i[0]].title)
    
    
```

```python
recommend('Titanic')
```
